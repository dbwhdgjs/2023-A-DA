## figsize()

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(5,2))

plt.rc('font', family= 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False


plt.plot([-2,-1,0,1,2,3],[20,30,15,40,45,35],label='선 1')
plt.plot([-2,-1,0,1,2,3],[-25,10,35,20,15,25],label='선 2')
         

plt.title('선 그래프 예제')
plt.xlabel('x축')
plt.ylabel('y축')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))

plt.rc('font', family= 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

x=np.linspace(-2*np.pi,2*np.pi,200)
plt.plot(x,np.sin(x)*10)
plt.plot(x,x**2)
plt.plot(x,2*x**2-3*x)
         
plt.grid(True)
plt.title('다양한 그래프')
plt.show()
```

# boxplot()

```python
import matplotlib.pyplot as plt
import random

result = []
for i in range(13):
    result.append(random.randint(1,1000))
print(sorted(result))

plt.boxplot(result)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

result1 = []
result2 = []
for i in range(13):
    result1.append(random.randint(1,500))
    result2.append(random.randint(501,1000))


plt.boxplot(result1)
plt.boxplot(result2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

result1 = []
result2 = []
for i in range(13):
    result1.append(random.randint(1,500))
    result2.append(random.randint(501,1000))


plt.boxplot([result1, result2])

plt.show()
```

## plt.figure(figsize=(m,n))

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)

f1 = plt.figure(figsize=(10,2))
plt.plot(np.random.randn(100))

plt.title("figure size : (10, 2)")
plt.show()
```

```python
ax1 = plt.subplot(2,1,1)
plt.plot(np.random.randint(1,10,10),'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(np.arange(10),np.arange(10)*2,'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig,axes = plt.subplots(2,2)

axes[0,0].plot([1,2,3,4])
axes[0,0].set_title("Subplot 1")

axes[0,1].plot([4,3,2,1])
axes[0,1].set_title("Subplot 2")

axes[1,0].hist([1,2,2,3,3,3,4,4,4,4],bins=5)
axes[1,0].set_title("Subplot 3")

axes[1,1].scatter([1,2,3,4],[4,3,2,1])
axes[1,1].set_title("Subplot 4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig,axes = plt.subplots(2,2)

axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title("axes 1")
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title("axes 2")
axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title("axes 3")
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title("axes 4")


plt.tight_layout()
plt.show()
```

# fig.add_subplot()

```python
fig = plt.figure(figsize = (7,5))
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

```python
fig = plt.figure(figsize = (10,8))
ax2 = fig.add_subplot(2,2,2)
plt.plot(np.random.randn(50).cumsum(),'g--')
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50),'m--')
```

```python
fig, axes= plt.subplots(2,2,sharex=True, sharey=True,figsize=(8,6))
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=.05)
```
